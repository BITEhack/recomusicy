### imports

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
```

### load datasets

```python
songsdata_df = pd.read_csv('muse_v3.csv')
songsdata_df = songsdata_df[['track', 'valence_tags', 'arousal_tags']]
display(songsdata_df)

songsliryc_df = pd.read_csv('tcc_ceds_music.csv')
display(songsliryc_df)
```

### Make titles lowercase

```python
songsdata_df['track'] = songsdata_df['track'].apply(lambda x : x.lower())
songsliryc_df['track_name'] = songsliryc_df['track_name'].apply(lambda x : x.lower())
```

### merge datasets

```python
tracks = [row['track'] for _, row in songsdata_df[['track']].iterrows()]

tracks_com = []

for _, track in songsliryc_df[['track_name']].iterrows():
    track = track['track_name']
    if track in tracks:
        tracks_com.append(track)
        
print(len(tracks_com))
```

```python
songsdata_df = songsdata_df.loc[songsdata_df['track'].isin(tracks_com)]
songsliryc_df = songsliryc_df.loc[songsliryc_df['track_name'].isin(tracks_com)]
```

```python
artist = []
for track in tracks_com:
    for _, row in songsliryc_df.iterrows():
        if row['track_name'] == track:
            artist.append(row['artist_name'])
            break
print(len(artist))

genres = []
for track in tracks_com:
    for _, row in songsliryc_df.iterrows():
        if row['track_name'] == track:
            artist.append(genres['genre'])
            break
print(len(genres))

lyrics = []
for track in tracks_com:
    for _, row in songsliryc_df.iterrows():
        if row['track_name'] == track:
            lyrics.append(row['lyrics'])
            break
print(len(lyrics))

valence = []
for track in tracks_com:
    for _, row in songsdata_df.iterrows():
        if row['track'] == track:
            valence.append(row['valence_tags'])
            break
print(len(valence))

arousal = []
for track in tracks_com:
    for _, row in songsdata_df.iterrows():
        if row['track'] == track:
            arousal.append(row['arousal_tags'])
            break
print(len(arousal))

data = {'Artist': artist,
        'Genre': genres,
        'Title': tracks_com, 
        'Lyric': lyrics,
        'Valence': valence,
        'Arousal': arousal} 


df = pd.DataFrame(data)
```

### save

```python
df.to_csv('songslirycs.csv', index=False)
```
