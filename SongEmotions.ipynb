### imports

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
```

### load datasets

```python
songsdata_df = pd.read_csv('muse_v3.csv')
songsdata_df = songsdata_df[['track', 'valence_tags', 'arousal_tags']]
display(songsdata_df)

songsliryc_df = pd.read_csv('Songs.csv')
display(songsliryc_df)
```

### Make titles lowercase

```python
songsdata_df['track'] = songsdata_df['track'].apply(lambda x : x.lower())
songsliryc_df['Title'] = songsliryc_df['Title'].apply(lambda x : x.lower())
```

### merge datasets

```python
tracks = [row['track'] for _, row in songsdata_df[['track']].iterrows()]

tracks_com = []

for _, track in songsliryc_df[['Title']].iterrows():
    track = track['Title']
    if track in tracks:
        tracks_com.append(track)
        
print(len(tracks_com))
```

```python
songsdata_df = songsdata_df.loc[songsdata_df['track'].isin(tracks_com)]
```

```python
lyrics = []
for track in tracks_com:
    for _, row in songsliryc_df.iterrows():
        if row['Title'] == track:
            lyrics.append(row['Lyrics'])
            break
print(len(lyrics))

valence = []
for track in tracks_com:
    for _, row in songsdata_df.iterrows():
        if row['track'] == track:
            valence.append(row['valence_tags'])
            break
print(len(valence))

arousal = []
for track in tracks_com:
    for _, row in songsdata_df.iterrows():
        if row['track'] == track:
            arousal.append(row['arousal_tags'])
            break
print(len(arousal))

data = {'Title': tracks_com, 
        'Lyric': lyrics,
        'Valence': valence,
        'Arousal': arousal} 


df = pd.DataFrame(data)
```

```python
df.to_csv('songslirycs.csv', index=False)
```
